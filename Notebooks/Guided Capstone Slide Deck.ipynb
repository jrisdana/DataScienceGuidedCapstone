### Problem Identification

Big Mountain Ski resort, located in Montana near Glacier National Park, recently installed a new chair lift to help increase the distribution of visitors across the mountain. The new lift increases seasonal operating costs by $1,540,000.

The resort currently prices tickets based on the market average of ski resort ticket prices. The client wants a pricing strategy that is informed by their facilities. Management seeks to create an investment strategy that maximizes the value of their equipment.

**Our goal:** Generate a price range for Big Mountain Resort based on the importance of the facilities they provide to visitors.

**Our data:** We are given a dataset with 330 ski resorts with features detailing the facilities offered and ticket prices for Weekdays and Weekends.

### Recommendation and key findings

Based on the given data, we recommend Big Mountain Resort **increases** its Weekday and Weekend ticket price.

### Summary and Conclusion

- Given the resort today, Big Mountain is undercharging consumers. The resort should **increase its ticket price** to the modelled price (95.87 dollars) to maximize utility of its facilities. 


- Big Mountain can **shut down one run and not affect revenue.** If Big Mountain needs to shut down three runs, it can shut down an additional two runs and not lower its ticket price more than it did for shutting down the initial three runs. There should be further analysis with maintenance cost data to see if closing down runs can make the resort more profitable.


- Adding runs, increasing the vertical drop and installing chairlifts will increase the ticket price. Big Mountain already has more to offer in runs, drops and chairlifts than the average resort. 


- Increasing snow coverage by several acres does not affect ticket price or revenue. 


- Important to know how elastic different consumer segments are to a potential ticket price increase. Although Big Mountain Ski Resort is is undercharging for its facilities, it is still the most expensive resort in Montana.
